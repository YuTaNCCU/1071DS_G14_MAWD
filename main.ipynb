```python
#Loading Packages
import tabu_search_functions
from importlib import reload
reload(tabu_search_functions)#tabu_search_functions若有更新，會重新載入
import time
import functions as functions
import numpy as np
```

```python
#Tabu Search + random起始法
TimeRecord=[] #一次優化耗時紀錄
ScoreRecord=[] #一次優化分數紀錄
Schedule_optimized=[] #一次優化結果List紀錄

for i in range(10):
    print('\n\n----------------RUN', i+1 ,'----------------')
    start_time = time.time()
    
    #Tabu Search Main Function:
    TSRun = tabu_search_functions.TabuSearch(InitialMethod='random',  _NumOfNeighbor=50, _NumOfSWAP=5, _NumOfTabuList=10, _MaxIteration=100)
    
    TimeRecord.append(time.time() - start_time)
    ScoreRecord.append(TSRun[0])
    Schedule_optimized.append(TSRun[1])
```

```python
# 展示結果

for i in range(len(ScoreRecord)):
    print('第',i,'輪分數(0~1000分): %.2f' %ScoreRecord[i], '  第',i,'輪耗時（秒）: %.2f' %TimeRecord[i])
print('此10輪最佳解為：\n')
functions.ListToSchedule(Schedule_optimized[np.argmax(ScoreRecord)])
```
